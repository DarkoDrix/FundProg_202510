````python
import random

def user_money():
    money = float(input("¿Cuánto dinero desea ingresar a la máquina? $: "))
    print("Haz ingresado $", money)
    return money

def select_words():
    palabras = ["Cerezas", "Naranjas", "Ciruelas", "Campanas", "Melones", "Barras"]
    w1 = random.choice(palabras)
    w2 = random.choice(palabras)
    w3 = random.choice(palabras)
    print(w1, "|", w2, "|", w3)
    return w1, w2, w3

def prize_logic(w1, w2, w3, money):
    if w1 == w2 == w3:
        ganancia = money * 3
        print("¡¡¡HAS ACERTADO TODAS!!!", ganancia)
    elif w1 == w2 or w1 == w3 or w2 == w3:
        ganancia = money * 2
        print("¡Felicidades, le has atinado a dos!", ganancia)
    else:
        ganancia = 0
        print("Que lastima, no ganaste nada :c", ganancia)
    return ganancia

# Aquí empieza a ejecutarse el código
print("Bienvenido a la máquina tragamonedas")
total_ganado = 0

while True:
    money = user_money()
    w1, w2, w3 = select_words()
    ganancia = prize_logic(w1, w2, w3, money)
    total_ganado += ganancia

    start = input("Do you want to play again? (Y/N): ").upper()
    if start == "N":
        print("\n Thanks for playing. Ganancias totales: $", total_ganado)
        break
    elif start != "Y":
        print("No recibimos una respuesta correcta. Fin del juego.")
        print("\n Thanks for playing. Ganancias totales: $", total_ganado)
        break
````

````python
# Estructuras para guardar datos
cuenta = {
    'username': None,
    'password': None,
    'money': 0,
    'historial': []
}

# Funciones
def create_account():
    username = input("Ingrese un nombre de usuario: ")
    password = input("Ingrese una contraseña: ")
    cuenta['username'] = username
    cuenta['password'] = password
    cuenta['money'] = 0
    cuenta['historial'] = []
    print("Cuenta creada exitosamente.\n")

def login_account():
    username_login = input("Ingrese su nombre de usuario: ")
    password_login = input("Ingrese su contraseña: ")
    if username_log == cuenta['username'] and password_log == cuenta['password']:
        print(f"Bienvenido, {username_log}.\n")
        return True
    else:
        print("Nombre de usuario o contraseña incorrectos.\n")
        return False

def depositar():
    try:
        cantidad = int(input("Ingrese una cantidad de dinero para depositar: "))
        if cantidad > 0:
            cuenta['money'] += cantidad
            cuenta['historial'].append(f"Depósito: +${cantidad}")
            print(f"Depósito exitoso. Nuevo saldo: ${cuenta['money']}\n")
        else:
            print("La cantidad debe ser mayor que 0.\n")
    except ValueError:
        print("Entrada inválida. Ingrese un número.\n")

def saldo_actual():
    print(f"Su saldo actual es: ${cuenta['money']}\n")

def retirar():
    try:
        cantidad = int(input("Ingrese una cantidad de dinero para retirar: "))
        if cantidad <= 0:
            print("La cantidad debe ser mayor que 0.\n")
        elif cantidad > cuenta['money']:
            print("Fondos insuficientes.\n")
        else:
            cuenta['money'] -= cantidad
            cuenta['historial'].append(f"Retiro: -${cantidad}")
            print(f"Retiro exitoso. Nuevo saldo: ${cuenta['money']}\n")
    except ValueError:
        print("Entrada inválida. Ingrese un número.\n")

def transacciones():
    if cuenta['historial']:
        print("Historial de transacciones:")
        for transaccion in cuenta['historial']:
            print(f"- {transaccion}")
    else:
        print("No hay transacciones registradas.\n")
    print()

# Programa principal
print("Bienvenido al Cajero Automático.")

# Lógica para crear e iniciar sesión
while True:
    if cuenta['username'] is None:
        deseo = input("No hay cuenta creada. ¿Desea crear una cuenta? (si/no): ").lower()
        if deseo == "si":
            create_account()
        else:
            print("Debe crear una cuenta para continuar.\n")
    else:
        deseo = input("¿Desea iniciar sesión? (si/no): ").lower()
        if deseo == "si":
            if login_account():
                break
        else:
            print("Debe iniciar sesión para continuar.\n")

# Menú de operaciones
while True:
    print("--- Menú de opciones ---")
    print("1. Depositar dinero")
    print("2. Consultar saldo")
    print("3. Retirar dinero")
    print("4. Ver historial de transacciones")
    print("5. Cerrar sesión")

    opcion = input("Seleccione una opción (1-5): ")
    print()

    if opcion == "1":
        depositar()
    elif opcion == "2":
        saldo_actual()
    elif opcion == "3":
        retirar()
    elif opcion == "4":
        transacciones()
    elif opcion == "5":
        print("Sesión cerrada. Gracias por usar el cajero automático.\n")
        break
    else:
        print("Opción no válida. Intente de nuevo.\n")
````

````python
record_list = []

inventario = {
    "ladrillo": {
        "id": 1,
        "buy_prize": 1200,
        "sell_prize": 1500,
        "quantity": 1000,
        "type": "unidad"
    },
    "Cemento Gris": {
        "id": 2,
        "buy_prize": 28000,
        "sell_prize": 30000,
        "quantity": 200,
        "type": "saco"
    },
    "Varilla Corrugada": {
        "id": 3,
        "buy_prize": 33000,
        "sell_prize": 35000,
        "quantity": 150,
        "type": "unidad"
    },
    "Malla Electro Soldada": {
        "id": 4,
        "buy_prize": 42000,
        "sell_prize": 45000,
        "quantity": 75,
        "type": "unidad"
    },
  }

usuarios = {
    "Ren Tao": {
        "current_user": False,
        "id": 1,
        "contraseña": "1234_4321",
        "rol": "admin"
    },
    "Ioh Asakur": {
        "current_user": False,
        "id": 2,
        "contraseña": "abc_cba",
        "rol": "vendedor"
    },
    "Chocolove McDonnell": {
        "current_user": False,
        "id": 3,
        "contraseña": 11110000,
        "rol": "vendedor"
    }
}

def search_product():
    print("\n--- Búsqueda de Productos ---")
    print("1. Buscar producto por nombre")
    print("2. Ver todo el inventario")
    opcion = int(input("Seleccione una opción: "))

    if opcion == 1:
        nombre = input("Ingrese el nombre del producto a buscar: ")
        producto = inventario.get(nombre)
        if producto:
            print(f"\nProducto encontrado: {nombre}")
            print(f"ID: {producto['id']}")
            print(f"Precio de compra: ${producto['buy_prize']}")
            print(f"Precio de venta: ${producto['sell_prize']}")
            print(f"Cantidad disponible: {producto['quantity']}")
            print(f"Tipo de unidad: {producto['type']}")
        else:
            print("Producto no encontrado en el inventario.")
    elif opcion == 2:
        print("\n--- Inventario completo ---")
        for nombre, datos in inventario.items():
            print(f"\nNombre: {nombre}")
            print(f"ID: {datos['id']}")
            print(f"Precio de compra: ${datos['buy_prize']}")
            print(f"Precio de venta: ${datos['sell_prize']}")
            print(f"Cantidad disponible: {datos['quantity']}")
            print(f"Tipo de unidad: {datos['type']}")
    else:
        print("Opción inválida.")

def add_record(tipo, id_producto, cantidad, total):
    if tipo == "buy":
        total = -abs(total)  # Compra(negativo)
    elif tipo == "sell":
        total = abs(total)   # Venta(positivo)

    record = {
        "tipo": tipo,
        "id": id_producto,
        "cantidad": cantidad,
        "total": total
    }
    record_list.append(record)


def vender_producto():
    print("\n--- Venta de Productos ---")
    nombre = input("Ingrese el nombre del producto a vender: ")

    if nombre not in inventario:
        print("Producto no encontrado.")
        return

    cantidad = int(input("Cantidad a vender: "))
    disponible = inventario[nombre]["quantity"]

    if cantidad > disponible:
        print("No hay suficiente inventario.")
        return

    precio_unitario = inventario[nombre]["sell_prize"]
    total_venta = precio_unitario * cantidad
    inventario[nombre]["quantity"] -= cantidad

    id_producto = inventario[nombre]["id"]

    # Registrar venta
    add_record("sell", id_producto, cantidad, total_venta)

    print(f"Venta realizada: {cantidad} unidades de '{nombre}' por ${total_venta}")

def add_product():
    print("Ingrese los datos del nuevo producto: ")
    nombre_producto = input("Nombre del producto: ")

    if nombre_producto in inventario:
        cantidad_extra = int(input("El producto ya existe. ¿Cuántas unidades desea agregar?: "))
        inventario[nombre_producto]["quantity"] += cantidad_extra

        # Registrar como compra
        precio_compra = inventario[nombre_producto]["buy_prize"]
        total = cantidad_extra * precio_compra
        id_producto = inventario[nombre_producto]["id"]
        add_record("buy", id_producto, cantidad_extra, total)

        print("Cantidad actualizada exitosamente.")
    else:
        nuevo_id = int(input("ID del producto: "))
        buy_price = int(input("Precio de compra: "))
        sell_price = int(input("Precio de venta: "))
        quantity = int(input("Cantidad disponible: "))
        tipo = input("Tipo de unidad (unidad/saco/etc): ")

        inventario[nombre_producto] = {
            "id": nuevo_id,
            "buy_prize": buy_price,
            "sell_prize": sell_price,
            "quantity": quantity,
            "type": tipo
        }

        # Registrar como compra
        total = quantity * buy_price
        add_record("buy", nuevo_id, quantity, total)

        print("Producto agregado exitosamente.")


def summary_sales():
    print("\n--- Resumen de Movimientos ---")
    if not record_list:
        print("No hay movimientos registrados.")
        return

    total_balance = 0
    for record in record_list:
        tipo = record['tipo']
        id_prod = record['id']
        cantidad = record['cantidad']
        total = record['total']

        print(f"{tipo.upper()} - ID: {id_prod} | Cantidad: {cantidad} | Total: ${total}")
        total_balance += total

    print(f"\n💰 BALANCE FINAL: ${total_balance}")

def admin_options():
    while True:
        print("\nBienvenido administrador")
        print("---------------------------")
        print("¿Qué le gustaría hacer?")
        print("1. Registrar un nuevo usuario")
        print("2. Añadir productos")
        print("3. Buscar productos")
        print("4. Resumen de ventas")
        print("5. Cerrar sesión")
        print("---------------------------")
        admin_choice = int(input("Ingrese una opción: "))
        if admin_choice == 1:
            register_user()
        elif admin_choice == 2:
            add_product()
        elif admin_choice == 3:
            search_product()
        elif admin_choice == 4:
            summary_sales()
        elif admin_choice == 5:
            print("Sesión cerrada.\n")
            break
        else:
            print("Opción inválida")

def seller_options(nombre_usuario):
    while True:
        print(f"\nBienvenido vendedor: {nombre_usuario}")
        print("---------------------------")
        print("1. Buscar productos")
        print("2. Vender productos")
        print("3. Cerrar sesión")
        print("---------------------------")
        choice = int(input("Ingrese una opción: "))

        if choice == 1:
            search_product()
        elif choice == 2:
            vender_producto()
        elif choice == 3:
            print("Sesión cerrada.\n")
            break
        else:
            print("Opción inválida")

def register_user():
    print("Ingrese los datos del nuevo usuario: ")
    new_user_name = input("Nombre de usuario: ")

    # Validar que no exista ya
    if new_user_name in usuarios:
        print("Error: el nombre de usuario ya existe.")
        return

    new_user_id = int(input("ID del usuario: "))
    new_user_password = input("Contraseña: ")
    new_user_rol = input("Rol (admin o vendedor): ").lower()

    # Verificar que el ID no esté repetido
    for datos in usuarios.values():
        if datos["id"] == new_user_id:
            print("Error: el ID ya está en uso.")
            return

    usuarios[new_user_name] = {
        "current_user": False,
        "id": new_user_id,
        "contraseña": new_user_password,
        "rol": new_user_rol
    }

    print(f"Usuario '{new_user_name}' registrado exitosamente.")

def login():
    password = input("Ingrese su contraseña: \n")
    id_user = int(input("Ingrese su ID: \n"))

    for nombre, datos in usuarios.items():
        if id_user == datos["id"] and str(password) == str(datos["contraseña"]):
            datos["current_user"] = True
            rol = datos["rol"]
            if rol == "admin":
                admin_options()
            else:
                seller_options(nombre)
            return
    print("Usuario o contraseña incorrectos")

def show_main_menu():
    print("Bienvenido a la tienda Shaman Tools")
    print("----------------------------------")
    print("¿Qué desea hacer?")
    print("1. Iniciar sesión")
    print("2. Salir")
    print("----------------------------------\n")
    user_choice = int(input("Ingrese una opción: "))
    if user_choice == 1:
        login()
    elif user_choice == 2:
        print("Gracias por su visita\n")
    return user_choice

while True:
  user_choice = show_main_menu()
  if user_choice == 2:
      break
````
